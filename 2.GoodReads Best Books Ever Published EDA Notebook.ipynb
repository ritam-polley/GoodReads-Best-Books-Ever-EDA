Reasons for the removal of the following columns:
1. Series: 
 The series column indicates whether a book belongs to a series and, if so, its position in that series. While this information provides context, it doesn't directly contribute to numerical factors like rating, price, or numRatings, which are more indicative of sales and popularity.
 When a book is bought or rated highly by readers and critiques the don't just look at the series only but also they need to check other additional information like the name of the author, series genre,etc.
 Series only affects sales and ratings of a book rarely if the book is a from a very reputed series eg. The Harry Potter Series, The Hunger Games Series, Sherlock Holmes, The Witcher Series etc. to name a few.
 So, the column "Series" is not important in our EDA process.


2. Description:
 Description contains text that summarized the overview of the story. The main purpose is to create an intrigue in the readers' minds before buying the book or reading. But this is of a very little impact because there are a very less number of readers who depend on the description for reading the book.     There are many other factors that come into play eg. Author, Initial Ratings, Pages, etc.
 So, the column "Description" is not important in our EDA process.
 

3. Characters:
 The characters column lists the key individuals in the book's story, offering insight into the narrative or series continuity. While this information is useful for storytelling or literary analysis, it holds minimal value in an EDA focused on numerical attributes like sales, ratings, or popularity. Additionally, the data in this column often overlaps with other attributes, such as the description, series, or title, making it redundant for the intended analysis. Since characters do not directly correlate with a book's sales performance or ratings, this column can be removed.
So, the column "Characters" is not important in our EDA process.


4. Edition: 
 The Edition holds a minimal value in the evaluation of sales of a book or the rating of the book because readers don't usually care about the edition of the book/novel. They buy it mostly for its contents and literary value.
 So, the column "Edition" is not important in our EDA process.


5. Setting:
The "Setting" column provides information about the geographical or temporal location of a book's story, such as "1850s England" for *Sherlock Holmes* or the cold, isolated landscapes of *Nordic thrillers*. While the setting can contribute to a book's atmosphere and appeal, its role in determining sales, ratings, or overall popularity is minimal. Readers often infer the setting from the book’s genre or title without relying on an explicit column. For instance, the gothic horror genre is inherently associated with eerie, isolated settings, while the *Sherlock Holmes* series is recognized for its Victorian London backdrop.
Moreover, the setting does not independently influence a book's performance; instead, it complements broader factors like the plot, characters, and genre. For example, a Nordic thriller’s frigid setting is compelling because of how it enhances the mystery and suspense, but it is the genre itself that primarily attracts readers. Similarly, a historical novel set in medieval England might captivate fans of historical fiction, but it is the genre and writing style, not the specific setting, that drive sales or ratings. Since the setting often overlaps with insights already captured by the series, title, or genre columns, it becomes redundant and minimally impactful in an EDA focused on numerical attributes or popularity trends. 
 So, the column "Setting" is not important in our EDA process.
 

6. Publihser:
 The publisher column indicates the publishing house responsible for producing and distributing the book. While this information provides context about the book's origin, it is minimally important in an EDA focused on determining sales, ratings, or popularity. Readers rarely choose books based on the publisher alone; instead, they prioritize factors such as the author, genre, ratings, or word-of-mouth recommendations.
 In some cases, a renowned publisher might lend credibility or visibility to a book. For example, a publication from Penguin Random House or HarperCollins could suggest quality or widespread availability. However, this influence is indirect and often overshadowed by more impactful factors like the book's content, marketing, or reviews. Furthermore, the publisher column often contains a wide variety of unique values, leading to high cardinality, which adds noise rather than meaningful insights to the analysis. So, the column "Publihser" is not important in our EDA process.
 

7. Cover Image:
The coverImg column contains URLs or references to the book's cover images, which serve primarily as a visual identifier or marketing tool. While an attractive cover may initially draw a reader’s attention, it has minimal direct impact on measurable metrics such as ratings, sales, or popularity. Readers typically evaluate books based on their author, genre, reviews, or word-of-mouth recommendations rather than the cover image alone. Furthermore, in the context of EDA, the coverImg column often introduces unnecessary complexity as it does not contribute to numerical or categorical analysis. So, the column "coverImg" is not important in our EDA process.


8. Book Format:
The bookFormat column specifies the physical or digital form of the book, such as hardcover, paperback, or eBook. While format can influence purchasing decisions to some extent—certain readers may prefer one format over another—it is not a dominant factor in determining a book’s overall popularity or rating. Additionally, sales data and ratings in this dataset are aggregated across all formats, making the individual format information less relevant. So, the column "BookFormat" is not important in our EDA process.


9. First Publish Date: 
The firstPublishDate column indicates the date when a book was originally published. While this information provides historical context, its impact on current sales or ratings is indirect. Older books may have higher ratings due to their established reputation, or newer books might gain traction through recent marketing efforts. However, this trend is better captured by more dynamic attributes like rating, numRatings, or price. Additionally, the firstPublishDate column often suffers from missing or inconsistent data, which could complicate the analysis without providing substantial insights. So, the column "firstPublishDate" is not important in our EDA process.


10. Awards: 
People don't tend to buy any book by looking at the number of awards it has recieved generally, also awards are the subsequent results of ratings by readers and media houses. So, the column "awards" is not important in our EDA process.

The graph showcases the average number of ratings (numRatings) for books published before and after the after the advent of internet on 1995. 2. Shift in Reading Behavior:
Prior to 1995, most ratings or reviews were confined to traditional platforms such as newspapers, magazines, or academic journals. After 1995, the internet democratized this process, enabling a wider audience to participate.
3. Global Reach:
The internet allowed books to reach a global audience. This wider reach likely contributed to the increase in the number of ratings for books published post-1995.
4. Platform Accessibility:
Online platforms made it easier for people to discover new books through personalized recommendations and social proof (ratings and reviews). This may have particularly benefited newer publications.
5. Cultural Shifts:
The graph could also highlight changing cultural trends in reading habits, where readers became more active participants in sharing their opinions online.

```python
#Filtering books published before and after 1995
before_1995 = datacopy[datacopy['publishDate'] < 1995]
after_1995 = datacopy[datacopy['publishDate'] >= 1995]
#Calculating the mean numRatings and counting the number of books for both groups
mean_numratings_before = before_1995['numRatings'].mean()
mean_numratings_after = after_1995['numRatings'].mean()
count_before = before_1995.shape[0]
count_after = after_1995.shape[0]

#Preparing data for the bar chart
labels = ['Before 1995', 'After 1995']
mean_numratings_values = [mean_numratings_before, mean_numratings_after]
count_values = [count_before, count_after]

#figure and subplots
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.4
r1 = np.arange(len(labels))
r2 = [x + bar_width for x in r1]

#Plotting the mean numRatings
ax.bar(r1, mean_numratings_values, color='pink', width=bar_width, edgecolor='grey', label='Mean numRatings')
#Plotting the number of books
ax.bar(r2, count_values, color='skyblue', width=bar_width, edgecolor='grey', label='Number of Books')

#labels and title
ax.set_xlabel('Publish Date Group', fontsize=12)
ax.set_ylabel('Value', fontsize=12)
ax.set_title('Comparison of Mean numRatings and Number of Books by Publish Date', fontsize=14)
ax.set_xticks([r + bar_width/2 for r in range(len(labels))])
ax.set_xticklabels(labels)

#legend
ax.legend()

# Display the chart
plt.show()
```

2. Correlation Values:

Correlation values range from -1 to 1:

1 means a perfect positive relationship.

-1 means a perfect negative relationship.

0 means no relationship.

Positive values indicate that as one factor increases, the other factor tends to increase.

Negative values indicate that as one factor increases, the other factor tends to decrease.

3. Strong Positive Correlations:

For example, "total5stars" and "total4stars" have a strong positive correlation, suggesting that books with many 5-star ratings often have many 4-star ratings as well.

4. Weak or Negative Correlations:

Factors like "total1stars" and "rating" might have a negative correlation, indicating that books with higher ratings tend to have fewer 1-star ratings.

Some factors might show weak correlations, meaning they don't strongly affect each other.

5. Interpreting the Heatmap:

The color intensity helps quickly identify the strength of the relationships:

Darker colors (close to red or blue) indicate stronger correlations.

Lighter colors indicate weaker correlations.
